```julia
positionSelector = SpatialVirtualSpecies.ExponentialPosSelector(10)
function testPosSelector(posSelector)
    x = zeros(51,51)
    idx = CartesianIndex(25,25)
    rng = MersenneTwister()
    for i in 1:10500
        newXY = SpatialVirtualSpecies.selectColonisedCoordinate(posSelector,idx,rng)
        if newXY[2]>=1 && newXY[2] <= 51 && newXY[1] >=1 && newXY[1]<=51
            x[newXY[1],newXY[2]]+=1
        end
    end
    return x
end
a = testPosSelector(positionSelector)
PyPlot.gray()
imshow(a)
colorbar()
```

```julia
positionSelector = SpatialVirtualSpecies.UniformPosSelector(10)
a = testPosSelector(positionSelector)
PyPlot.gray()
imshow(a)
colorbar()
```
